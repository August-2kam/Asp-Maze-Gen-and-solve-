```python
sns.set_theme(style="whitegrid", context="talk")

runtime_stats = (
    df.groupby(["algorithm", "size"]) ["wall_time_sec"]
      .agg(mean="mean", std="std", median="median")
      .reset_index()
)

plt.figure(figsize=(12, 7))
for algo in runtime_stats["algorithm"].unique():
    sub = runtime_stats[runtime_stats["algorithm"] == algo].sort_values("size")
    plt.errorbar(
        sub["size"],
        sub["mean"],
        yerr=sub["std"],
        marker="o",
        linewidth=2,
        capsize=4,
        label=algo,
    )

plt.yscale("log")
plt.xlabel("Graph Size")
plt.ylabel("Wall Time (seconds, log scale)")
plt.title("Runtime vs Size (All Algorithms, Mean ± SD)")
plt.legend(title="Algorithm", ncol=2)
plt.tight_layout()
plt.show()

c_stats = runtime_stats[runtime_stats["algorithm"].str.lower() != "asp"]

plt.figure(figsize=(12, 7))
for algo in c_stats["algorithm"].unique():
    sub = c_stats[c_stats["algorithm"] == algo].sort_values("size")
    plt.errorbar(
        sub["size"],
        sub["mean"],
        yerr=sub["std"],
        marker="o",
        linewidth=2,
        capsize=4,
        label=algo,
    )

plt.xlabel("Graph Size")
plt.ylabel("Wall Time (seconds)")
plt.title("Runtime vs Size (Traditional Algorithms Only, Mean ± SD)")
plt.legend(title="Algorithm")
plt.tight_layout()
plt.show()
```
